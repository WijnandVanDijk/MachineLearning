```python
class Neuron:
    
    def __init__(self, bias, weight):
        self.weight = weight
        self.bias = bias
        
        self.output = 0
        self.error = 0
        self.output = 0
        self.learning_rate = 0.1
        
    def activate(self, x):
        return 1/(1+2.71828**(-x))
    
    def get_output(self, _input):
        som = 0
        for i in range(len(_input)):
            som += _input[i] * self.weight[i]
        x = som + self.bias
        self.output = self.activate(x)
        return self.output
    
    def error(self, target):
        e = self.output * (1-self.output) * -(target - self.output)
        self.error = e
        return e
    
    def gradient(self):
        return self.output * self.error
    
    def update(self, _input, target, runs):

        for y in range(runs):
            y = self.get_output(_input)
            print(y)

            error = target - y

            for i in range(len(self.weight)):
                delta = self.learning_rate * error * _input[i]
                self.weight[i] = delta + self.weight[i]
            
            delta_bias = self.learning_rate * error
            self.bias = self.bias + delta_bias
            
    def error_hidden(self, prev_neurons, neuron_count):
        weighted_errors = []
        for neuron in prev_neurons.neurons:
            weighted_error = neuron.error * neuron.weight[neuron_count]
            weighted_errors.append(weighted_error)
        som_errors = sum(weighted_errors)

        sigmoid = self.output * (1 - self.output)

        self.error = sigmoid * som_errors

        return self.error
        
    def __str__(self):
        return ('Bias: {}' + ' Weight: {}').format(self.bias, self.weight)
```

```python
class NeuronLayer:
    
    def __init__(self):
        self.neurons = []
        
    def setter(self, gates):
        self.neurons.append(gates)
        
    def getter(self):
        return self.neurons
    
    def output_layer(self, _input):
        next_input = []
        for neuron in self.neurons:
            next_input.append(neuron.get_output(_input))
        return next_input
    
    def __str__(self):
        return ('Aantal neurons in laag: {}'.format(len(self.neurons)))
```

```python
class NeuronNetwork:
    
    def __init__(self):
        self.layers = []
        
    def setter_layers(self, aantal_layers):
        self.layers.append(aantal_layers)
        
    def feed_forward(self, _input):
        #outputs_netwerk = []
        for layer in self.layers:
            next_input = layer.output_layer(_input)
            _input = next_input 
        return _input  
    
  
#    def train(self, inputs, target):
#        teller = 0
#        while teller < 10: # epochs
#            for i in range(target):
#                losses = []
#                for y in range(target):
#                    output = self.feed_forward(inputs)
#                    loss = output - target
#                    losses.append(loss)
#                output = self.feed_forward(inputs)
#            teller += 1   
```

### AND gate
De verwachte output is 0 0 0 1

```python
AND = Neuron(random.randint(0, 5), [5, -2])
for i in range(250):
    print(AND)
    AND.update([0,0], 0, 1)
    AND.update([0,1], 0, 1)
    AND.update([1,0], 0, 1)
    AND.update([1,1], 1, 1)
```

```python
layer1 = NeuronLayer()
layer1.setter(AND)
print(layer1)
```

```python
Netwerk = NeuronNetwork()
Netwerk.setter_layers(layer1)
```

```python
print(Netwerk.feed_forward([0, 0]))
print(Netwerk.feed_forward([0, 1]))
print(Netwerk.feed_forward([1, 0]))
print(Netwerk.feed_forward([1, 1]))
```
